```python
# What type of email servers are there and if they exist.
emails = df.EMAIL.unique().tolist()
servidor_mail = []
for email in emails:
    try:
        if email.find('@') != -1 and email[email.find('@') + 1:] not in servidor_mail:
            servidor_mail.append(email[email.find('@') + 1:])
        elif email.find('@') == -1:
            print(f'wrong email: {email}')
    except:
        print(email)
servidor_mail
```

There are two types of email providers: one is wrong 'com' and the other is correct gmail.

```python
# For each unique value of mail server how many clients are there.
emails = df.EMAIL.unique().tolist()
com = 0
gmail = 0
for email in emails:
    try:
        if email.find('@') != -1:
            if email[email.find('@') + 1:] == 'com':
                com += 1
            else:
                gmail += 1
        elif email.find('@') == -1:
            print(f'wrong email: {email}')
    except:
        print(email)
print(f'com has {com} values and gmail has {gmail} values.')
```

7 clients do not have email, 63 have the wrong email and 890 have gmail.com email.

```python
# Emails with the letter ñ do not exist.
count = 0
count_gmail = 0
for mail in df[df.EMAIL.notnull()].EMAIL.unique():
    if 'ñ' in mail:
        count += 1
        if 'gmail' in mail:
            count_gmail += 1
print(f'There are {count} emails with the letter ñ and {count_gmail} gmail emails with the letter ñ.')
```

```python
# Existing unique cp services.
new_cp = []
for cp in df.DOM_SERVICIO_CP.unique():
    if type(cp) == str:
        new_cp.append(int(cp))
    else:
        new_cp.append(cp)
sorted(set(new_cp))
```

Correct typographical errors of DOM_SERVICIO_CP

```python
# Correct typographical error (tildes) of provinces.
df.DOM_SERVICIO_PROVINCIA = df.DOM_SERVICIO_PROVINCIA.apply(lambda x:
                            'ALMERÍA' if x == 'ALMERIA'
                            else ( 'MÁLAGA' if x == 'MALAGA'
                            else ( 'CÓRDOBA' if x == 'CÓRDOBA'
                            else x)))
```

```python
# Correct error from str to integer.
df.DOM_SERVICIO_CP = df.DOM_SERVICIO_CP.apply(lambda x: int(x) if type(x) == str else x)
```

```python
# Provinces that share service cp.
cp_keys = {}
for cp in df.DOM_SERVICIO_CP:
    cp_keys[cp] = []
for index, row in df.iterrows():
    cp_keys[row['DOM_SERVICIO_CP']].append(row['DOM_SERVICIO_PROVINCIA'])

for cp in cp_keys:
    cp_keys[cp] = set(cp_keys[cp])
for key in cp_keys:
    if len(cp_keys[key]) > 1:
        print(key, cp_keys[key])
```
